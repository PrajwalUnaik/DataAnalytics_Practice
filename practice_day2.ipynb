```python
from scipy.stats import norm

data = df['sepal length (cm)']
mean = data.mean()
std = data.std()
lower_bound = 5.0
upper_bound = 6.0

probability = norm.cdf(upper_bound, loc=mean, scale=std) - norm.cdf(lower_bound, loc=mean, scale=std)

print(f"Probability between {lower_bound} and {upper_bound}: {probability}")
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target] # Add species names as a column
df.head()
```

```python
df.describe()  # Basic descriptive statistics
df.groupby('species').mean() # Group by species and calculate mean
df.groupby('species').std()  # Group by species and calculate standard deviation
```

```python
   import matplotlib.pyplot as plt
   import seaborn as sns

   # Histogram
   plt.figure(figsize=(8, 6))
   sns.histplot(data=df, x='sepal length (cm)', hue='species', kde=True)
   plt.title('Distribution of Sepal Length by Species')
   plt.show()

   # Box plot
   plt.figure(figsize=(8, 6))
   sns.boxplot(data=df, x='species', y='sepal length (cm)')
   plt.title('Sepal Length Variation across Species')
   plt.show()

   # Scatter plot
   plt.figure(figsize=(8, 6))
   sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
   plt.title('Relationship between Sepal Length and Width')
   plt.show()
```
